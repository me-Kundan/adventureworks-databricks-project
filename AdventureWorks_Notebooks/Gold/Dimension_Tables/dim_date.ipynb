Import Required Libraries and Spark SQL Functions

```python
import os
from pyspark.sql.functions import *
```

Define Gold and Silver Layer Base Paths

```python
silver_base = "/Volumes/adventure_works_lakehouse/adventure_works/lakehouse/silver"
gold_base   = "/Volumes/adventure_works_lakehouse/adventure_works/lakehouse/gold"
```

Read and Transform Silver Calendar Data

```python
df_dim_date = spark.read.format("delta")\
                .load(f"{silver_base}/silver_adventureworks_calendar")\
                .withColumn("DateKey", date_format(col("Date"), "yyyyMMdd").cast("int"))\
                .withColumn("MonthName", date_format(col("Date"), "MMMM"))\
                .withColumn("Quarter", quarter(col("Date")))\
                .select(
                    "DateKey",
                    col("Date"),
                    "Month",
                    "MonthName",
                    "Quarter",
                    "Year"
                )
```

```python
df_dim_date.limit(10).display()
```

Write Transformed data into Gold Layer

```python
df_dim_date.write.format("delta")\
            .mode("overwrite")\
            .save(f"{gold_base}/dim_date")
```
