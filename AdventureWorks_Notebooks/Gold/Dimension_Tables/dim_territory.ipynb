Import Required Libraries and Spark SQL Functions

```python
import os
from pyspark.sql.functions import *
```

Define Gold and Silver Layer Base Paths

```python
silver_base = "/Volumes/adventure_works_lakehouse/adventure_works/lakehouse/silver"
gold_base   = "/Volumes/adventure_works_lakehouse/adventure_works/lakehouse/gold"
```

Read and Prepare Territory Dimension Data

```python
df_dim_territory = (
    spark.read.format("delta")
    .load(f"{silver_base}/silver_adventureworks_territories")
    .select(
        col("SalesTerritoryKey").alias("TerritoryKey"),
        "Region",
        "Country",
        "Continent"
    )
)
```

```python
df_dim_territory.limit(10).display()
```

Write Territory Dimension Data to Gold Layer

```python
df_dim_territory.write.format("delta")\
            .mode("overwrite")\
            .save(f"{gold_base}/dim_territory")
```
